```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import ensemble
from sklearn import cross_validation
from sklearn.metrics import roc_auc_score as auc
import time
```

```python
train = pd.read_csv('train.csv/train.csv')
test = pd.read_csv('test/test.csv')
```

```python
plt.rcParams['figure.figsize'] = (10,10)
```

```python
##Remove columns with zero standard deviation
zero_std_col = []
for column in train.columns:
    if train[column].std()==0:
        zero_std_col.append(column)

train.drop(zero_std_col,axis=1,inplace=True)        
```

```python
##Remove identical columns
col_identical_col = []
for i in range(0,len(train.columns)-1):
    for j in range(i+1,len(train.columns)):
        if np.array_equal(train[train.columns[i]].values,train[train.columns[j]].values)==True:
            col_identical_col.append(i)


col_to_remove = []
[col_to_remove.append(train.columns[i]) for i in col_identical_col]
train.drop(col_to_remove,axis=1,inplace=True)
```

```python
trainlabels = train['TARGET']
trainfeatures = train.drop(['ID','TARGET'],axis=1)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import auc
from sklearn.ensemble import gradient_boosting
```

```python
gbm_classifier = ensemble.GradientBoostingClassifier(n_estimators=20,max_features=1,max_depth=3,min_samples_leaf=100
                                                            ,subsample=0.65,learning_rate=0.1,loss='deviance',random_state=1)
```

```python
x_train,x_valid,y_train,y_valid = train_test_split(trainfeatures,trainlabels,test_size=0.5,random_state=1)
```

```python
featuretable = pd.DataFrame(index=range(len(x_train.columns)),columns=['Feature','AUC'])
for i,feature in enumerate(x_train.columns):
    traininputfeature = x_train[feature].values.reshape(-1,1)
    validinputfeature = x_valid[feature].values.reshape(-1,1)
    gbm_classifier.fit(traininputfeature,y_train)
    valid_auc = auc(y_valid,gbm_classifier.predict_proba(validinputfeature)[:,1],reorder=True)
    featuretable.ix[i,'Feature'] = feature
    featuretable.ix[i,'AUC'] = valid_auc
```
